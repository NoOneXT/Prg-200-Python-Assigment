# Question 1 - Small Shop Billing and Inventory System

```python
def process_order(inventory, cart):
    """
    Process customer order, calculate bill, update stock.
    
    Parameters:
    inventory: dict with items containing price and stock
    cart: dict with items and quantities ordered
    """
    print("=" * 40)
    print("BILL")
    print("=" * 40)
    
    total_bill = 0
    
    # Process each item in cart
    for item, quantity in cart.items():
        if item in inventory:
            available_stock = inventory[item]["stock"]
            price = inventory[item]["price"]
            
            if quantity <= available_stock:
                item_total = price * quantity
                total_bill += item_total
                inventory[item]["stock"] -= quantity
                print(f"{item} x{quantity} = NPR {item_total}")
            else:
                print(f"Sorry, not enough stock for {item}")
        else:
            print(f"Sorry, {item} is not available in inventory")
    
    print("=" * 40)
    print(f"Grand Total: NPR {total_bill}")
    print("=" * 40)
    print("\nUpdated stock:")
    
    for item, details in inventory.items():
        print(f"{item} = {details['stock']}", end="  ")
    print("\n")
    
    return total_bill


# Given data
inventory = {
    "rice": {"price": 120, "stock": 20},
    "milk": {"price": 90, "stock": 10},
    "bread": {"price": 60, "stock": 15},
    "eggs": {"price": 15, "stock": 30}
}

cart = {
    "rice": 2,
    "milk": 3,
    "eggs": 12
}

# Run the function
process_order(inventory, cart)
```

# Question 2 - Water Level Alert System (Koshi River)

```python
def check_water_level(location, level_metres):
    """
    Check water level and return alert status.
    
    Parameters:
    location: name of the sensor location
    level_metres: water level in metres
    
    Returns:
    alert status string
    """
    if level_metres < 3:
        status = "Safe"
    elif 3 <= level_metres <= 5:
        status = "Warning - Alert nearby villages"
    else:  # level_metres > 5
        status = "DANGER - Evacuate immediately!"
    
    print(f"{location} ({level_metres} m): {status}")
    return status


# Given data
sensors = [
    ("Chatara", 2.8),
    ("Tribeni Ghat", 5.4),
    ("Koshi Barrage", 4.1),
    ("Sunsari Bridge", 1.9),
    ("Saptakoshi Camp", 6.0)
]

# Check all sensors
print("KOSHI RIVER WATER LEVEL ALERT SYSTEM")
print("=" * 45)
for location, level in sensors:
    check_water_level(location, level)
```

# Question 3 - Date Converter for Nepal Bank System (BS ↔ AD)

```python
def convert_date(date_str, from_cal, to_cal):
    """
    Convert date between Bikram Sambat (BS) and Gregorian (AD) calendars.
    
    Parameters:
    date_str: date in format 'YYYY-MM-DD'
    from_cal: 'BS' or 'AD'
    to_cal: 'BS' or 'AD'
    
    Returns:
    converted date string
    """
    if from_cal == to_cal:
        return date_str
    
    # Split the date
    year, month, day = date_str.split("-")
    year = int(year)
    
    # Convert based on direction
    if from_cal == "AD" and to_cal == "BS":
        converted_year = year + 56
        converted_month = month
        converted_day = day
    elif from_cal == "BS" and to_cal == "AD":
        converted_year = year - 56
        converted_month = month
        converted_day = day
    else:
        return date_str
    
    return f"{converted_year:04d}-{converted_month}-{converted_day}"


def format_nepali_date(date_str):
    """Convert ISO date to Nepali format with month names."""
    bs_months = [
        "Baisakh", "Jestha", "Ashadh", "Shrawan", "Bhadra", "Ashwin",
        "Kartik", "Mangsir", "Poush", "Magh", "Falgun", "Chaitra"
    ]
    year, month, day = date_str.split("-")
    month_num = int(month)
    day_num = int(day)
    
    # Nepali date formatting
    if 1 <= day_num <= 9:
        day_str = f"{day_num}th"
    elif day_num == 1:
        day_str = "1st"
    elif day_num == 2:
        day_str = "2nd"
    elif day_num == 3:
        day_str = "3rd"
    else:
        day_str = f"{day_num}th"
    
    return f"{day_str} {bs_months[month_num-1]}, {year}"


# Test with customers
customers = [
    {"name": "Ramesh Thapa", "date": "1985-06-24", "cal": "AD", "need": "BS", "style": "full"},
    {"name": "Sunita Karki", "date": "2055-09-10", "cal": "BS", "need": "AD", "style": "iso"},
    {"name": "Bikash Rai", "date": "1998-11-30", "cal": "AD", "need": "BS", "style": "nepali"},
    {"name": "Anjali Gurung", "date": "2040-01-05", "cal": "BS", "need": "AD", "style": "full"},
]

print("DATE CONVERSION RESULTS")
print("=" * 55)

for customer in customers:
    converted = convert_date(customer["date"], customer["cal"], customer["need"])
    
    if customer["style"] == "nepali" and customer["need"] == "BS":
        converted_display = format_nepali_date(converted)
    elif customer["style"] == "full" and customer["need"] == "BS":
        converted_display = format_nepali_date(converted) + " BS"
    elif customer["style"] == "full" and customer["need"] == "AD":
        converted_display = f"{converted} AD"
    else:  # iso style
        converted_display = f"{converted} {customer['need']}"
    
    print(f"{customer['name']} | Original: {customer['date']} {customer['cal']} | Converted: {converted_display}")
```

# Question 4 - Word Frequency Counter

```python
def word_frequency(text):
    """
    Count frequency of each word in text and return top 3 most frequent.
    
    Parameters:
    text: string of text to analyze
    
    Returns:
    list of tuples (word, count) for top 3 words
    """
    # Convert to lowercase and split into words
    words = text.lower().split()
    
    # Remove punctuation from each word
    cleaned_words = []
    for word in words:
        # Remove punctuation at start and end of word
        clean_word = word.strip('.,!?;:()"\'')
        cleaned_words.append(clean_word)
    
    # Count frequencies using dictionary
    freq_dict = {}
    for word in cleaned_words:
        if word:  # ignore empty strings
            freq_dict[word] = freq_dict.get(word, 0) + 1
    
    # Sort by frequency (descending) and get top 3
    sorted_words = sorted(freq_dict.items(), key=lambda x: x[1], reverse=True)
    top_3 = sorted_words[:3]
    
    # Print results
    print("Top 3 words:")
    for word, count in top_3:
        print(f"  {word} - {count} times")
    
    return top_3


# Test text
text = """
Nepali is a beautiful country. Nepal has Mount Everest. 
Everest is the highest mountain in the world. Many tourists 
visit Nepal every year to see Everest and other mountains. 
Nepal is known for its mountains and natural beauty. 
"""

print("WORD FREQUENCY ANALYSIS")
print("=" * 35)
word_frequency(text)
```

# Question 5 - Simple ATM Simulator

```python
def atm(account_id, pin, action, amount=0):
    """
    Simple ATM function for cooperative bank.
    
    Parameters:
    account_id: account identifier
    pin: 4-digit PIN
    action: 'balance', 'deposit', or 'withdraw'
    amount: amount for deposit/withdraw (default 0)
    """
    accounts = {
        "A001": {"name": "Ramesh Thapa", "balance": 15000, "pin": "1234"},
        "A002": {"name": "Sunita Karki", "balance": 8500, "pin": "5678"},
        "A003": {"name": "Bikash Rai", "balance": 22000, "pin": "9012"}
    }
    
    print("\n" + "=" * 50)
    
    # Check if account exists
    if account_id not in accounts:
        print(f"Account '{account_id}' not found")
        return
    
    account = accounts[account_id]
    
    # Check PIN
    if pin != account["pin"]:
        print(f"Incorrect PIN for account {account_id}")
        return
    
    print(f"Welcome, {account['name']}!")
    
    # Perform action
    if action == "balance":
        print(f"Your current balance: NPR {account['balance']}")
    
    elif action == "deposit":
        account["balance"] += amount
        print(f"Deposited NPR {amount}")
        print(f"New balance: NPR {account['balance']}")
    
    elif action == "withdraw":
        if amount <= account["balance"]:
            account["balance"] -= amount
            print(f"Withdrew NPR {amount}")
            print(f"Remaining balance: NPR {account['balance']}")
        else:
            print(f"Insufficient funds. Your balance is NPR {account['balance']}")
    
    else:
        print(f"Invalid action: {action}")
    
    print("=" * 50)


# Test the ATM function
print("ATM SIMULATOR - Lalitpur Cooperative Bank")
print("=" * 50)

# Test case 1: Balance check
atm("A001", "1234", "balance")

# Test case 2: Wrong PIN
atm("A002", "0000", "withdraw", 2000)

# Test case 3: Deposit
atm("A002", "5678", "deposit", 3000)

# Test case 4: Insufficient funds
atm("A003", "9012", "withdraw", 25000)

# Test case 5: Account not found
atm("A004", "1111", "balance")
```
