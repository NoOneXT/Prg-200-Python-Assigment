# Question 1 - Bank Account Manager

```python
class BankAccount:
    def __init__(self, name, account_number, balance=0):
        self.name = name
        self.account_number = account_number
        self.balance = balance
    
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited NPR {amount}. New balance: NPR {self.balance}")
        else:
            print("Deposit amount must be positive.")
    
    def withdraw(self, amount):
        if amount > self.balance:
            print(f"Insufficient funds! Cannot withdraw NPR {amount}. Current balance: NPR {self.balance}")
        elif amount <= 0:
            print("Withdrawal amount must be positive.")
        else:
            self.balance -= amount
            print(f"Withdrew NPR {amount}. New balance: NPR {self.balance}")
    
    def get_balance(self):
        print(f"Account: {self.name} (A/C: {self.account_number}) | Current Balance: NPR {self.balance}")


# Test the class
print("=" * 50)
print("BANK ACCOUNT MANAGER")
print("=" * 50)

# Create accounts
accounts = [
    BankAccount("Ramesh Thapa", "A001", 5000),
    BankAccount("Sunita Karki", "A002", 0),
    BankAccount("Bikash Rai", "A003", 12000)
]

# Perform transactions
print("\n--- Initial Balances ---")
for acc in accounts:
    acc.get_balance()

print("\n--- Depositing NPR 3000 into A002 (Sunita Karki) ---")
accounts[1].deposit(3000)

print("\n--- Attempting to withdraw NPR 15000 from A003 (Bikash Rai) ---")
accounts[2].withdraw(15000)

print("\n--- Withdrawing NPR 2000 from A001 (Ramesh Thapa) ---")
accounts[0].withdraw(2000)

print("\n--- Final Balances ---")
for acc in accounts:
    acc.get_balance()
```

# Question 2 - Student Report Card

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
    
    def average(self):
        return sum(self.marks) / len(self.marks)
    
    def grade(self):
        avg = self.average()
        if avg >= 80:
            return "A"
        elif avg >= 65:
            return "B"
        elif avg >= 50:
            return "C"
        elif avg >= 40:
            return "D"
        else:
            return "F"
    
    def display(self):
        avg = self.average()
        grd = self.grade()
        status = "Pass" if avg >= 40 else "Fail"
        print(f"Student: {self.name}")
        print(f"  Marks: {self.marks}")
        print(f"  Average: {avg:.2f}")
        print(f"  Grade: {grd}")
        print(f"  Status: {status}")
        print("-" * 30)


# Test the class
print("=" * 50)
print("STUDENT REPORT CARD")
print("=" * 50)
print()

students = [
    Student("Aarav", [78, 85, 60, 90, 72]),
    Student("Sita", [45, 50, 38, 60, 55]),
    Student("Bishal", [30, 25, 40, 35, 28]),
    Student("Priya", [90, 88, 95, 92, 87])
]

for student in students:
    student.display()
```

# Question 3 - Food Delivery App


```python
class DeliveryPartner:
    def __init__(self, name, partner_id, deliveries):
        self.name = name
        self.partner_id = partner_id
        self.deliveries = deliveries
    
    def total_earning(self):
        # To be overridden by child classes
        return 0
    
    def display(self):
        print(f"Partner: {self.name} ({self.partner_id})")
        print(f"  Deliveries: {self.deliveries}")
        print(f"  Total Earnings: NPR {self.total_earning():.2f}")
        print("-" * 35)


class BikeRider(DeliveryPartner):
    def __init__(self, name, partner_id, deliveries, km_travelled):
        super().__init__(name, partner_id, deliveries)
        self.km_travelled = km_travelled
    
    def total_earning(self):
        # NPR 80 per delivery + NPR 5 per km
        return (self.deliveries * 80) + (self.km_travelled * 5)


class Walker(DeliveryPartner):
    def __init__(self, name, partner_id, deliveries, rainy_deliveries):
        super().__init__(name, partner_id, deliveries)
        self.rainy_deliveries = rainy_deliveries
    
    def total_earning(self):
        # NPR 60 per delivery + NPR 50 per rainy delivery
        return (self.deliveries * 60) + (self.rainy_deliveries * 50)


class CarDriver(DeliveryPartner):
    def __init__(self, name, partner_id, deliveries, fuel_cost):
        super().__init__(name, partner_id, deliveries)
        self.fuel_cost = fuel_cost
    
    def total_earning(self):
        # NPR 120 per delivery - fuel cost
        return (self.deliveries * 120) - self.fuel_cost


# Test the classes
print("=" * 50)
print("FOOD DELIVERY APP")
print("=" * 50)
print()

partners = [
    BikeRider("Santosh Rai", "B-01", 15, 42),
    Walker("Kabita Maharjan", "W-01", 18, 5),
    CarDriver("Roshan KC", "C-01", 20, 850)
]

# Display all partners
print("--- ALL DELIVERY PARTNERS ---")
for partner in partners:
    partner.display()

# Find partner with highest earning
print("\n--- HIGHEST EARNING PARTNER ---")
highest_earner = max(partners, key=lambda p: p.total_earning())
print(f" {highest_earner.name} ({highest_earner.partner_id}) earns the most: NPR {highest_earner.total_earning():.2f}")
```

# Question 4 - Bus Ticket Booking (Sajha Yatayat)

```python
class Bus:
    def __init__(self, route, total_seats):
        self.route = route
        self.total_seats = total_seats
        self.booked_seats = {}  # Dictionary: seat_number -> passenger_name
    
    def book_seat(self, seat_number, passenger_name):
        if seat_number < 1 or seat_number > self.total_seats:
            print(f"Invalid seat number! Seat {seat_number} does not exist. (1-{self.total_seats})")
        elif seat_number in self.booked_seats:
            print(f"Seat {seat_number} already booked by {self.booked_seats[seat_number]}")
        else:
            self.booked_seats[seat_number] = passenger_name
            print(f"✓ Seat {seat_number} booked successfully for {passenger_name}")
    
    def available_seats(self):
        return self.total_seats - len(self.booked_seats)
    
    def passenger_list(self):
        if not self.booked_seats:
            print("No seats booked yet.")
            return
        
        print(f"\n{'='*40}")
        print(f"PASSENGER LIST - {self.route}")
        print(f"{'='*40}")
        print(f"{'Seat No':<10} {'Passenger Name':<20}")
        print(f"{'-'*10} {'-'*20}")
        
        for seat in sorted(self.booked_seats.keys()):
            print(f"{seat:<10} {self.booked_seats[seat]:<20}")
        
        print(f"{'='*40}")
        print(f"Total Booked: {len(self.booked_seats)} | Available: {self.available_seats()}")


# Test the class
print("=" * 50)
print("SAJHA YATAYAT - BUS TICKET BOOKING")
print("=" * 50)
print()

# Create bus
bus = Bus("Kathmandu - Pokhara", 10)
print(f" Route: {bus.route}")
print(f" Total Seats: {bus.total_seats}")
print()

# Process bookings
bookings = [
    (3, "Ramila Shrestha"),
    (7, "Deepak Gurung"),
    (3, "Anita Rai"),      # Duplicate seat
    (1, "Prakash Magar"),
    (7, "Suman Tamang")    # Duplicate seat
]

print("--- PROCESSING BOOKINGS ---")
for seat, name in bookings:
    bus.book_seat(seat, name)
    print(f"   Available seats left: {bus.available_seats()}")

# Final status
print(f"\n--- FINAL STATUS ---")
print(f"Available seats: {bus.available_seats()}")

# Display passenger list
bus.passenger_list()
```
