# Question 1 - Temperature Logger (math module)

```python
import math

# Global variable
station_name = "Kathmandu Weather Station"

def get_average(temps):
    """Returns mean temperature"""
    if not temps:
        return 0
    return sum(temps) / len(temps)

def get_deviation(temps):
    """Returns standard deviation using math.sqrt"""
    mean = get_average(temps)  # Local variable - only accessible inside this function
    if not temps:
        return 0
    variance = sum((t - mean) ** 2 for t in temps) / len(temps)
    return math.sqrt(variance)

def get_summary(temps):
    """Prints min, max, average, and deviation"""
    print(f"=" * 50)
    print(f"{station_name}")
    print(f"=" * 50)
    print(f"Temperature Data: {temps}")
    print(f"-" * 50)
    print(f"Minimum Temperature: {min(temps)}°C")
    print(f"Maximum Temperature: {max(temps)}°C")
    print(f"Average Temperature: {get_average(temps):.2f}°C")
    print(f"Standard Deviation: {get_deviation(temps):.4f}°C")
    print(f"=" * 50)


# Test the functions
temperatures = [18.4, 22.1, 15.7, 29.3, 11.8, 25.6, 19.2]

print("\n" + "=" * 50)
print("TEMPERATURE LOGGER")
print("=" * 50)

get_summary(temperatures)

# Demonstrating scope
print("\n--- SCOPE DEMONSTRATION ---")
print("Inside get_deviation(), 'mean' is a LOCAL variable.")
print("Trying to access 'mean' outside the function causes a NameError:")
try:
    print(mean)  # This will cause NameError because mean is not defined in global scope
except NameError as e:
    print(f"   NameError: {e}")
    print("   This happens because 'mean' exists only within get_deviation() function.")
```

# Question 2 — Bill Splitter (random module)

```python
import random

# Set seed for consistent results
random.seed(42)

# Given data
friends = ["Ramesh", "Sunita", "Bikash", "Anjali", "Dipak"]
total_bill = 3750

def split_bill(friends, total):
    """Returns amount each person pays (equal split)"""
    if not friends:
        return 0
    return total / len(friends)

def pick_lucky(friends):
    """Uses random.choice to pick one person"""
    if not friends:
        return None
    return random.choice(friends)

def final_summary(friends, total):
    """Prints each person's share and the lucky person who pays extra NPR 50"""
    print("=" * 50)
    print("RESTAURANT BILL SPLITTER - THAMEL")
    print("=" * 50)
    
    # Calculate base share
    base_share = split_bill(friends, total)
    
    # Pick lucky person
    lucky_person = pick_lucky(friends)
    
    # Local variable for lucky person's total
    lucky_total = base_share + 50
    
    print(f"\nTotal Bill: NPR {total}")
    print(f"Number of Friends: {len(friends)}")
    print(f"Base Share per Person: NPR {base_share:.2f}")
    
    print(f"\n--- FINAL BREAKDOWN ---")
    for person in friends:
        if person == lucky_person:
            print(f"  {person}: NPR {lucky_total:.2f} (includes NPR 50 lucky tax) ")
        else:
            print(f"  {person}: NPR {base_share:.2f}")
    
    print(f"\n--- SUMMARY ---")
    print(f" Lucky Person: {lucky_person} (pays extra NPR 50)")
    print(f" {lucky_person}'s total: NPR {lucky_total:.2f}")
    print("=" * 50)
# Test the function
final_summary(friends, total_bill)
```

# Question 3 — Exam Scheduler (datetime module)

```python
import datetime

# Global variable
college_name = "Bhaktapur Multiple Campus"

def parse_date(date_str):
    """Converts string to datetime object"""
    # Remove any spaces from the date string
    clean_date = date_str.strip().replace(" ", "")
    return datetime.datetime.strptime(clean_date, "%Y-%m-%d")

def get_exam_date(start_str, days):
    """Returns start date + days as a string 'YYYY-MM-DD'"""
    start_date = parse_date(start_str)
    exam_date = start_date + datetime.timedelta(days=days)
    return exam_date.strftime("%Y-%m-%d")

def print_schedule(start_str, exams):
    """Prints each subject and its exam date"""
    print("=" * 60)
    print(f"{college_name}")
    print("=" * 60)
    print(f" Exam Schedule (Starting from: {start_str})")
    print("=" * 60)
    print(f"{'Subject':<25} {'Days from Start':<15} {'Exam Date':<12}")
    print("-" * 60)
    
    for subject, days in exams:
        exam_date = get_exam_date(start_str, days)
        print(f"{subject:<25} {days:<15} {exam_date:<12}")
    
    print("=" * 60)


# Test the functions
start_date = "2025-05-01"
# Fixed: Using proper list of tuples
exams = [
    ("Python Programming", 0),
    ("Data Structures", 3),
    ("Database Systems", 6),
    ("Computer Networks", 10),
    ("Mathematics", 14),
]

print_schedule(start_date, exams)
```

# Question 4 — Shopping Discount (own module)

```python
# Run this cell FIRST to create the discount.py module file
import os  

module_content = '''# discount.py - Custom module for discount calculations

# Global constant (13% VAT)
TAX_RATE = 0.13

def apply_discount(price, percent):
    """Returns price after discount"""
    if percent < 0 or percent > 100:
        return price  # Fixed indentation
    discount_amount = price * (percent / 100)
    return price - discount_amount

def apply_tax(price):
    """Returns price after adding TAX_RATE"""
    return price * (1 + TAX_RATE)

def final_price(price, discount_pct):
    """Applies discount first, then tax"""
    discounted_price = apply_discount(price, discount_pct)
    final = apply_tax(discounted_price)
    return final
'''

# Write the module to a file
with open('discount.py', 'w', encoding='utf-8') as f:
    f.write(module_content)

print("✓ discount.py module created successfully!")
print("✓ File location:", os.path.abspath('discount.py'))
```
