# Question 1 — ATM Withdrawal Validator 

```python
balance = float(input("Enter acccount balance: "))
daily_withdrawn = float(input("How much money was withdrawn today?"))
amount = float(input("Enter amount to withdraw: "))

if amount % 500 != 0:
    print("Invalid amount. Must be a multiple of NPR 500.")
elif balance < amount:
    print(f"Insufficient balance, you have NPR {balance} only.")
elif daily_withdrawn + amount > 50000:
    print("Daily withdrawal limit reached (NPR 50,000).")
else:
    print("Withdrawal successful.")
    balance = balance - amount
    print(f"Your current balance after withdrawal: NPR {balance}")
```

# Question 2 — Online Store Discount System

```python
total_amount = float(input("Enter Total Amount: "))
loyalty_mem = input("Are you a loyalty member? (y/n) ").lower()

if total_amount < 1000:
    final_amount = total_amount
elif total_amount < 5000:
    final_amount = total_amount * 0.95
elif total_amount < 15000:
    final_amount = total_amount * 0.90
else:
    final_amount = total_amount * 0.80

if loyalty_mem == 'y':
    final_amount = final_amount * 0.95

print(f"Final Amount: NPR {final_amount}")
```

# Question 3 — Inventory Restock Alert

```python
inventory = [
    {"item": "Rice", "stock": 5, "threshold": 10},
    {"item": "Eggs", "stock": 24, "threshold": 12},
    {"item": "Milk", "stock": 3, "threshold": 6},
    {"item": "Bread", "stock": 8, "threshold": 5},
    {"item": "Chicken", "stock": 0, "threshold": 4},
    {"item": "Cooking Oil", "stock": 2, "threshold": 3},
]

count = 0
for item in inventory:
    if item["stock"] < item["threshold"]:
        print(f"Item: {item["item"]}, Stock: {item["stock"]}, Threshold: {item["threshold"]}")
        print("-"* 50)
        count += 1
        
print(f"Items that need restocking: {count}")
```

# Question 4 — Password Strength Checker 

```python
passwords = ["hello", "Hello123", "H3ll0@World", "12345678", "MyP@ss!"]
special_chars = "!@#$%^&*"

for pwd in passwords:
    print(f"Password: {pwd}")
    missing = []

    if len(pwd) < 8:
        missing.append("at least 8 characters")
    if not any(c.isupper() for c in pwd):
        missing.append("uppercase letter")
    if not any(c.islower() for c in pwd):
        missing.append("lowercase letter")
    if not any(c.isdigit() for c in pwd):
        missing.append("digit")
    if not any(c in special_chars for c in pwd):
        missing.append(f"special character ({special_chars})")

    if missing:
        print("Weak password, missing:", ", ".join(missing))
    else:
        print("Strong password!")
    print("-" * 50)
```

# Question 5 — Taxi Fare Calculator 

```python
trips = [
    {"distance": 1.5, "hour": 14},
    {"distance": 5.0, "hour": 22},
    {"distance": 12.0, "hour": 3},
    {"distance": 8.5, "hour": 10},
    {"distance": 2.0, "hour": 23},
]

rate = 0
fare = 0

for trip in trips:
    fare = fare + 1
    if trip["distance"] <= 2:
        rate = 150
    elif trip["distance"] <= 10:
        rate = 150 + (trip["distance"] - 2) * 35
    else:
        rate = 150 + 8 * 35 + (trip["distance"] - 10) * 28
    if trip["hour"] < 5 or trip["hour"] >= 22:
        rate = rate * 1.1
    print(f"Fare {fare} Total: NPR {rate}")
```
